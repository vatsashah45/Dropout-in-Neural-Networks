```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
score = json.load(open('WorkingModel', 'r'))
plt.figure(1)
plt.plot(score['acc'])
plt.plot(score['val_acc'])
plt.title('Working model accuracy - MNIST')
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.ylim([0.0,1.0])
plt.savefig('WorkingModel.png', bbox_inches='tight')
plt.show()
```
