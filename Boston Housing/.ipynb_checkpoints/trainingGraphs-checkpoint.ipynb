```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
score = json.load(open('OverfittedModel', 'r'))
plt.figure(1)
plt.plot(score['mae'])
plt.plot(score['val_mae'])
plt.title('Overfitted model mean absolute error - Boston Housing')
plt.ylabel('mae')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.ylim([0.0,35])
plt.savefig('OverfittedModel.png', bbox_inches='tight')
plt.show()
```

```python
score = json.load(open('DropoutModel', 'r'))
plt.figure(1)
plt.plot(score['mae'])
plt.plot(score['val_mae'])
plt.title('Dropout model mean absolute error - Boston Housing')
plt.ylabel('mae')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.ylim([0.0,35])
plt.savefig('DropoutModel.png', bbox_inches='tight')
plt.show()
```
